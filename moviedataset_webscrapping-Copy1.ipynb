```python
from bs4 import BeautifulSoup
import requests
import numpy as np
import pandas as pd
```

```python
new_df = pd.DataFrame(columns=[['name','year','genre','description','director','stars']])
```

```python
for pages in range(1,9):
    print(pages)
    url = 'https://www.imdb.com/list/ls002939221/?sort=list_order,asc&st_dt=&mode=detail&page={}'.format(pages)
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    paragraphs = soup.find_all(class_='lister-item')
    
    df = pd.DataFrame(columns=[['name','year','genre','description','director','stars']])
    
    for i in range(0,len(paragraphs)):
        name = paragraphs[i].find(class_='lister-item-content').find('a').text
        year = paragraphs[i].find(class_='lister-item-content').find(class_='lister-item-year').text[1:-1]
        genre = paragraphs[i].find(class_='lister-item-content').find(class_='genre').text

        des = paragraphs[i].find(class_='lister-item-content').find_all('p')[1].text[1:]


        
        director = paragraphs[i].find(class_='lister-item-content').find_all('p')[2].find('a').text
        


        for j in range(1,len(paragraphs[0].find(class_='lister-item-content').find_all('p')[2].find_all('a'))):

            stars = paragraphs[i].find(class_='lister-item-content').find_all('p')[2].find_all('a')[1].text
            stars +=  ' ,' + paragraphs[i].find(class_='lister-item-content').find_all('p')[2].find_all('a')[2].text
            stars +=  ' ,' + paragraphs[i].find(class_='lister-item-content').find_all('p')[2].find_all('a')[3].text


        df.loc[i]=[name,year,genre,des,director,stars]
    
    new_df = new_df.append(df,ignore_index=True)
```
